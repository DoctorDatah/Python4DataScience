```python
fig = plt.figure()
subplot1 = fig.add_subplot(2,2, 1)
subplot2 = fig.add_subplot(2,2, 2)
subplot3 = fig.add_subplot(2,2, 3)

plt.plot(np.random.randn(50).cumsum(), 'k--')


_ = subplot1.hist(np.random.randn(100), bins=20, color='k', alpha=.3)

subplot2.scatter(np.arange(30),np.arange(30))
```

```python
fig, axes = plt.subplots(2,2, sharex=True, sharey = True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500) , bins=50, color='r', alpha =0.5)
        
plt.subplots_adjust(wspace = 0, hspace=0)
plt.legend
```

```python
plt.plot(np.random.randn(200).cumsum())
plt.xlim()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1, 1)


ax.plot(np.random.randn(1000).cumsum(), 'r', label = 'red')
ax.plot(np.random.randn(1000).cumsum(), 'k--', label = 'gray dash')
ax.plot(np.random.randn(1000).cumsum(), 'k.', label = 'gray dot')
ax.legend(loc='best')
```

Annotation and Drawing on subplot 

```python
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(np.random.randn(1000).cumsum(), 'r', label = 'red')
ax.annotate('Hell here', (2,10))
ax.arrow(0,0,xy=(10,10))

```
