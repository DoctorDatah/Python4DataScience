```python
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)


data = pd.read_csv('examples/spx.csv', index_col=0, parse_dates=True)
spx = data['SPX']

spx.plot(ax=ax, style = 'b-')

crisis_data = [
    (datetime(2007,10,11), 'peaknof the bull market'),
    (datetime(2008,3,12), 'Bear Stearns Fails '),
    (datetime(2008,9,15),'Lehman Bankrupcy')
    ]

for date, label in crisis_data:
    {
        ax.annotate(label, xy=(date, spx.asof(date) + 75),
                    xytext=(date,spx.asof(date) + 225),
                    arrowprops=dict(facecolor='black',headwidth=4,width=2,headlength=4),
                    horizontalalignment='left',
                    verticalalignment='top'
                   )
    }
    
ax.set_xlim([datetime(2007,1,1),datetime(2011,1,1) ])    
ax.set_ylim([600,1800])
_ = ax.set_title('Important dates in the 2008-2009 financial crisis')
```

### Annotate Arguments 

xy : (float, float)

    The point *(x,y)* to annotate.

xytext : (float, float), optional

    The position *(x,y)* to place the text at.
    If *None*, defaults to *xy*.
